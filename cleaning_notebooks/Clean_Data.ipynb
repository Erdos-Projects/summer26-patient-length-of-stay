```python
import ast

def num_codes(z):
    if isinstance(z, list):
        return len(z)

    if pd.isna(z):
        return 0

    if isinstance(z, str):
        z = z.replace("nan", "None")
        return len(ast.literal_eval(z))

    return 0
```

```python
def clean_data(file_name):
    df = pd.read_csv(file_name, low_memory= False)
    
    df = df.drop(columns = ['PAT_COUNTRY', 'PAT_STATE', 'PAT_COUNTY'])
    
    df['QUARTER'] = file_name[41:42]
    df['YEAR'] = file_name[43:47]
    df = df.drop(columns = 'DISCHARGE')
    
    nomi = pgeocode.Nominatim("us")

    df["zip_clean"] = (
        df["PAT_ZIP"]
        .astype(str)
        .str[:5]
        .str.zfill(5)
    )
    
    unique_zips = df["zip_clean"].dropna().unique()

    zip_lookup = nomi.query_postal_code(unique_zips)

    zip_lookup = zip_lookup[["postal_code", "latitude", "longitude"]]

    df = df.merge(
        zip_lookup,
        left_on="zip_clean",
        right_on="postal_code",
        how="left"
    )

    df = df.drop(columns = ['PAT_ZIP', 'zip_clean', 'postal_code'])
    df = df.rename(columns = {"latitude":"PAT_LATITUDE", "longitude" : "PAT_LONGITUDE"})
    
    df["zip_clean"] = (
        df["PROVIDER_ZIP"]
        .astype(str)
        .str[:5]
        .str.zfill(5)
    )
    
    unique_zips = df["zip_clean"].dropna().unique()

    zip_lookup = nomi.query_postal_code(unique_zips)

    zip_lookup = zip_lookup[["postal_code", "latitude", "longitude"]]

    df = df.merge(
        zip_lookup,
        left_on="zip_clean",
        right_on="postal_code",
        how="left"
    )
    
    df = df.drop(columns = ['PROVIDER_ZIP', 'zip_clean', 'postal_code'])
    df = df.rename(columns = {"latitude":"PROVIDER_LATITUDE", "longitude" : "PROVIDER_LONGITUDE"})
    
    df = df[df['PAT2PROV_DISTANCE'] > 0]
    
    df = df[df['TYPE_OF_ADMISSION'] != 9]
    
    df = df[df['SOURCE_OF_ADMISSION'] != "9"]
    
    return df
    
```
